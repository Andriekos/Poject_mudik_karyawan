```python
lokasi.head(10)

#Dibagi dua yaitu Jabodetabek dan Non-Jabodetabek
```

```python
tanggal = []

for t in lokasi['tanggal']:
    if t not in tanggal: tanggal.append(t)
        
jabodetabek = ['Jakarta Pusat', 'Jakarta Selatan', 'Jakarta Timur', 'Jakarta Barat',' Jakarta Utara',
              'Kabupaten Bogor', 'Kota Tangerang', 'Kota Tangerang Selatan', 'Kota Bogor','Kota Bekasi',
              'Kabunpaten Tangerang','Kabupaten Bekasi','Kota Depok'
              ]

karyawanLuarKota = []
karyawanDalamKota = []

for t in tanggal:
    cntLK = cntDK = 0
    
    for location in lokasi.lokasi_dominan[lokasi['tanggal']== t]:
        if location in jabodetabek: cntDK += 1
        else: cntLK += 1
            
    karyawanLuarKota.append(cntLK)
    karyawanDalamKota.append(cntDK)
```

```python
d1 = pd.DataFrame(karyawanLuarKota, columns = ['Jumlah Karyawan di Luar Kota'])
d2 = pd.DataFrame(karyawanDalamKota, columns = ['Jumlah Karyawan di Dalam Kota'])
d3 = pd.DataFrame(tanggal, columns = ['Tanggal'])
```

```python
data = pd.merge(d1,d2, left_index = True, right_index = True)
data_2 = pd.merge(d3, data, left_index = True, right_index = True)
```

```python
print(data_2)
print('------------------------------------------------')
print(data_2.describe())
```

```python
f, axes = plt.subplots(1,1,figsize=(20,8))
axes.plot(data_2['Tanggal'],data_2.iloc[:,1], label = 'Di luar Kota')
axes.plot(data_2['Tanggal'],data_2.iloc[:,2], label = 'Di dalam Kota')

plt.xticks(rotation=90)
plt.legend(loc='best')
plt.show()

axes.set(xlabel='Tanggal', ylabel='Jumlah Karyawan', title='Grafik keberadaan karyawan')
```

```python
from collections import Counter
```

```python
data_mudik = lokasi.iloc[1000:1900,:]
```

```python
lokasi.iloc[1000:1900,:].head()
```

```python
lokasi_mudik = []

for ID in range(1,101):
    lokasi_karyawan = list(data_mudik.lokasi_dominan[data_mudik.id == ID])
    modus = Counter(lokasi_karyawan).most_common(1)
    
    lokasi_mudik.append(modus[0][0])
    
ID = range(1,101)

df = list(zip(ID, lokasi_mudik))
data_lokasi_mudik = pd.DataFrame(df, columns= ['ID','Lokasi Mudik'])
```

```python
print(data_lokasi_mudik)
```

```python
# Perbandingan antara karyawan yang mudik dan yang tidak mudik
```

```python
jabodetabek = ['Jakarta Pusat', 'Jakarta Selatan', 'Jakarta Timur', 'Jakarta Barat',' Jakarta Utara',
              'Kabupaten Bogor', 'Kota Tangerang', 'Kota Tangerang Selatan', 'Kota Bogor','Kota Bekasi',
              'Kabunpaten Tangerang','Kabupaten Bekasi','Kota Depok'
              ]

persentase = [0.0, 0.0]

for l in data_lokasi_mudik['Lokasi Mudik']:
    if l in jabodetabek:
        persentase[1] += 1
    else:
        persentase[0] += 1
```

```python
labels = 'Mudik', 'Tidak Mudik'

f, axes = plt.subplots(1,1, figsize=(10,8))

patches, texts, autotext = axes.pie(persentase, explode=(0.1,0.1), labels= labels,
                                  autopct='%1.1f%%', startangle = 240, colors = ['gold','green']
                                  )

texts[0].set_fontsize(20)
texts[1].set_fontsize(20)

axes.set_title('Presentasi mudik karyawan', size=20)
axes.axis('equal')

plt.show()
```
